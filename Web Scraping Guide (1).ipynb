#  <font color=bleu>Method 2:

```python
import bs4
import requests
import pandas as pd
from bs4 import BeautifulSoup
```

```python
url='https://en.wikipedia.org/wiki/List_of_largest_companies_in_the_United_States_by_revenue'
response=requests.get (url)
soup= BeautifulSoup(response.text,'html.parser')
```

```python
table=soup.findAll('table')[1]
table
```

```python
titles = table.find_all('th')
titles
```

<font color=red>Obtain every title of columns with tag 'th': 

```python
headers = []
for i in titles :
 title = i.text.strip()
 headers.append(title)
```

```python
headers
```

```python
df = pd.DataFrame(columns = headers)
```

```python
df
```

```python
rows = table.find_all('tr')
```

```python
for row in rows [1:]:
    columns = row.find_all('td')
    column = [c.text.strip() for c in columns]

    length = len(df)
    df.loc[length] = column
```

```python
df
```

```python
df.to_csv("table.data.csv")
```

# <font color=redbold>Image Web Scraping

```python
import bs4
import requests
from bs4 import BeautifulSoup
```

```python
url='https://gazaishere.com/index.php/disaster-image-gallery/'
response=requests.get(url)
soup=BeautifulSoup(response.text, 'html.parser')
print(soup.prettify()[1:3000])
```

```python
images=soup.findAll('img')
```

```python
images
```

```python
images_url=[]
for image in images:
     src= image.get('src')
     images_url.append(src)

images_url
```

```python
# Download the images
for i, url in enumerate(images_url):
        response = requests.get(url)
        if response.status_code == 200:
            with open(f'image_{i}.jpg', 'wb') as f:
                f.write(response.content)
            print(f"Downloaded image {i}")
        else:
            print(f"Failed to download image {i}")
     
```

```python
for i, url in enumerate(images_url):
        response = requests.get(url)
        with open(f'image_{i}.jpg', 'wb') as f:
         f.write(response.content)
         print(f"Downloaded image {i}")
```
